```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import seaborn as sns
```

方程式を定義

```python
def equations(vars):
    a,b,c,d = vars
    eq1 = a - (0.25 * (5 + 0.7 * b) + 0.25 * 0.7 * c + 0.5 * 0.7 * a)
    eq2 = b - (0.5 * (5 + 0.7 * b) + 0.25 * 0.7 * d + 0.5 * 0.7 * a)
    eq3 = c - (0.25 * 0.7 * a + 0.25 * 0.7 * d + 0.5 * 0.7 * c)
    eq4 = d - (0.25 * (5 + 0.7 * b) + 0.25 * 0.7 * c + 0.5 * 0.7 * d)
    return [eq1, eq2, eq3, eq4]
```

初期値を設定

```python
initial_guess = [0,0,0,0]
```

方程式を解く

```python
solution = fsolve(equations, initial_guess)
```

結果を出力

```python
a,b,c,d = solution
print(f"a: {a:.2f}, b: {b:.2f}, c: {c:.2f}, d: {d:.2f}")
```

可視化

```python
answer = [[a,b],[c,d]]
plt.figure(figsize=(6,6))
plt.title("Small Markov Chain")
sns.heatmap(answer,annot=True,fmt=".2f",cbar=False,cmap="coolwarm")
plt.xticks([0.5,1.5],["C","D"],rotation=0)
plt.yticks([0.5,1.5],["A","C"],rotation=0)
plt.show()
```
